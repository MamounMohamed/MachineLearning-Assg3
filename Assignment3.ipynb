```python
import numpy as np 
import sklearn as sck
import sklearn.cluster as cluster
import matplotlib.pyplot as plt
import pandas as pd
from scipy.spatial import distance
```

# Question 1

```python
x=np.array([[2,4],[3,3],[3,4],[5,4],[5,6],[5,8],[6,4],[6,5],[6,7],[7,3],[7,4],[8,2],[9,4],[10,6],[10,7],[10,8],[11,5],[11,8],[12,7],[13,6],[13,7],[14,6],[15,4],[15,5]])
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
kmeans = cluster.KMeans(n_clusters=2,init='k-means++').fit(x)
centers = kmeans.cluster_centers_
labels = kmeans.predict(x)
plt.scatter(x[:,0],x[:,1], c=labels, cmap='rainbow')
plt.scatter(centers[:,0] ,centers[:,1],color='black')
```

```python
kmeans = cluster.KMeans(n_clusters=3,init='k-means++').fit(x)
centers=kmeans.cluster_centers_
labels = kmeans.predict(x)
plt.scatter(x[:,0],x[:,1],c=labels,cmap='rainbow')
plt.scatter(centers[:,0],centers[:,1],color='black')
```

```python
#the code below is just for calculating the matrix for internal measures
k=0 
y=[]
for element in x : 
    y.append([element ,labels[k]])
    k+=1
matrix=[] 
foo=[]
for i in range(3):
    for j in range (3):
              foo.append(0)
    matrix.append(foo)
    foo=[]

for i in range (len(y)-1):
    for j  in range (i+1,len(y)):
        dist = distance.euclidean(y[i][0], y[j][0])
        cluster_i = y[i][1] 
        cluster_j=y[j][1]
        matrix[cluster_i][cluster_j]+=dist
        if cluster_i!=cluster_j:
             matrix[cluster_j][cluster_i]+=dist
    
for i in matrix:
        for j in i: 
                print(j,end='  ')
        print()
```

```python
kmeans = cluster.KMeans(n_clusters=4,init='k-means++').fit(x)
centers=kmeans.cluster_centers_
labels = kmeans.predict(x)
plt.scatter(x[:,0],x[:,1],c=labels,cmap='rainbow')
plt.scatter(centers[:,0],centers[:,1],color='black')
```
